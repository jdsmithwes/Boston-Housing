Crime Data Split

Based on the correlation table above, the prediction that homes near the Charles River enjoy a lower rate of crime per capita than those that are further away. Additionally, the data supports the widely-held belief that as crime increases, the value of homes decreases. To filter the crime data, two dataframes can be created that detail home values in high crime areas versus low crime ones. High crime areas have per capita rates higher than the median of .26 and low crime areas will have a median below .26

```python
High = df['crim'] >=.26
Low = df['crim'] <= .26
```

```python
High_Crime = df[High]
Low_Crime = df[Low]
```

```python
High_Crime.describe()
```

```python
Low_Crime.describe()
```

The median home value for homes in low crime areas is $233,500 versus $190,000 in high crime areas which supports the hypothesis made earlier.

Quality Schools affect on Home Value

Another widely accepted belief is that better quality schools leads to higher home values. The median pupil to teacher ratio in the entire data set is 19 students per 1 teacher. High-quality schools will have a ratio below 19 and low quality schools will have a ratio above 19.

```python
Good = df['ptratio'] < 19
Bad = df['ptratio'] > 19
Good_Schools = df[Good]
Bad_Schools = df[Bad]
```

```python
print(Good_Schools.describe())
```

```python
print(Bad_Schools.describe())
```

As hypothesized, schools with higher pupil-teacher ratios have median home values of $184,000 while those with lower pupil-teacher ratios have a median home value of $240,000.
